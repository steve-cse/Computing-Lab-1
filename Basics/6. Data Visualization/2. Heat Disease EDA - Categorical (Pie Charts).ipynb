```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

## Displot for numerical data

```python
sns.displot(df['RestingBP'],kde=True)
```

```python
sns.displot(df['Cholesterol'],kde=True)
```

```python
sns.distplot(df['FastingBS'],kde=True,color='blue')
```

## Pie Charts for categorical data

```python
df.groupby('Sex').size()
```

```python
df.groupby('Sex').size().plot(kind='pie',autopct='%.f')
```

```python
df.groupby('ChestPainType').size().plot(kind='pie',autopct='%.1f')
```

```python
df.groupby('RestingECG').size().plot(kind='pie',autopct='%.1f')
```
