```python
import pandas as pd
import seaborn as sns
```

```python
df=pd.read_csv("heart.csv")
df.head()
```

```python
# Find correlation
df.corr()
# the more closer to 1 the better correlation
# -ve values show negative correlation
```

```python
sns.heatmap(df.corr())
```

```python
# MAXHR and Age has negative correlation
sns.jointplot(x='Age',y='MaxHR',data=df)
```

```python
# for zero correlation
sns.jointplot(x='Age',y='Age',data=df)
```

```python
sns.jointplot(x='HeartDisease',y='Oldpeak',data=df)
```

```python
sns.jointplot(x='HeartDisease',y='Cholesterol',data=df)
```

```python
# plot a regression line
sns.jointplot(x='Age',y='MaxHR',data=df,kind='reg')
```
