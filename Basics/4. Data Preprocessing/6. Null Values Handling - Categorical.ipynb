```python
import pandas as pd
# pip install scikit-learn
from sklearn.impute import SimpleImputer
import numpy as np
```

```python
df=pd.read_csv("Data.csv")
df
```

```python
# impute with most frequent value
imputer=SimpleImputer(missing_values=np.nan,strategy='most_frequent')
```

```python
# fetch salary and age columns as array
df.iloc[:,1:3].values
```

```python
imputer.fit(df.iloc[:,1:3].values)
# replace with impputed values
df.iloc[:,1:3] = imputer.transform(df.iloc[:,1:3].values)
```

```python
df
```

```python
# fill up country with most frequent values
df.iloc[:,:1]
```

```python
imputer.fit(df.iloc[:,:1].values)
df.iloc[:,:1]=imputer.transform(df.iloc[:,:1].values)
df
```

```python
imputer = SimpleImputer(missing_values=np.nan,strategy='most_frequent')
imputer.fit(df.iloc[:,3:].values)
df.iloc[:,3:] = imputer.transform(df.iloc[:,3:].values)
```
