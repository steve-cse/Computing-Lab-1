```python
import pandas as pd
```

```python
df = pd.read_csv("googleplaystore.csv")
df.head(5)
```

## Q1. Total number of apps in each category

```python
for name in df['Category'].unique():
    print(name)
```

```python
for i in df['Category']:
    print(i)
```

```python
categories = {}

for name in df['Category'].unique():
    ct = 0
    for i in df['Category']:
        if (i == name):
            ct += 1
    categories[name] = ct

print(categories)
```

## Q2. Total number of apps in each Type

```python
types={}
for names in df['Type'].unique():
    ct=0
    for i in df['Type']:
        if (i==names):
            ct+=1
    types[names]=ct
print(types)
```
