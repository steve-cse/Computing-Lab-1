```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
```

```python
df=pd.read_csv("googleplaystore.csv")
imputer=SimpleImputer(missing_values=np.nan,strategy='mean')
imputer.fit(df.iloc[:,2:3].values)
df.iloc[:,2:3]=imputer.transform(df.iloc[:,2:3].values)
df=df.dropna()
df.head()
```

## Q1. Name the apps that are in ART_AND_DESIGN with rating more than 4.5 in ascending order wrt Ratings

```python
df_pr=df[df['Category']=='ART_AND_DESIGN']
df_pr=df[df['Rating']>4.5]
df_pr.sort_values( by= 'Rating',ascending=True)
```

## Q2. Name the apps that are in ART_AND_DESIGN in descending order wrt Ratings

```python
df_pr=df[df['Category']=='ART_AND_DESIGN']
df_pr=df[df['Rating']>4.5]
df_pr.sort_values( by= 'Rating',ascending=False)
```

## Q3. Which category is having maximum average rating?

```python
df.groupby('Category').mean()['Rating'].sort_values(ascending=False)
```

## Q3. How many free apps are there in each category wrt type

```python
df_pr=df[df['Type']=='Free']
df_pr.groupby('Category').count()['Type'].sort_values(ascending=False)
```
