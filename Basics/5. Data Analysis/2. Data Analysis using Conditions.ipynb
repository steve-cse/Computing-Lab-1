```python
import numpy as np
import pandas as pd

from sklearn.impute import SimpleImputer
```

```python
df = pd.read_csv('googleplaystore.csv')

impute = SimpleImputer(missing_values = np.nan , strategy = 'mean')
impute.fit(df.iloc[ : , 2:3 ].values)
df.iloc[ : , 2:3 ] = impute.transform(df.iloc[ : , 2:3 ].values)

df = df.dropna()
```

```python
df[df['Category'] == 'BOOKS_AND_REFERENCE']
# we get the rows with category BOOKS_AND_REFERENCE
```

## Q1. How many free apps are there in ART_AND_DESIGN?

```python
df_processed=df[df['Category']=='ART_AND_DESIGN']
```

```python
df_processed[df_processed['Type']=='Free']
# 61 apps are free and the rows are displayed
```

## Q2. How many apps are there in ART_AND_DESIGN with rating more then 4.5?

```python
df_processed=df[df['Category']=='ART_AND_DESIGN']
df_processed=df_processed[df_processed['Rating']>4.5]
df_processed
```

## Q3. How many apps are there in FAMILY with rating more then 4.5 and Free?

```python
df_processed=df[df['Category']=='FAMILY']
df_processed=df_processed[df_processed['Rating']>4.5]
df_processed[df_processed['Type']=='Free']
# 314 apps 
```
